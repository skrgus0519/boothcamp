# BERT

```python
from transformers import TFBertForMaskedLM, AutoTokenizer
```

```python
# MODEL_NAME = 'bert-large-uncased'
MODEL_NAME = 'bert-base-uncased'
model = TFBertForMaskedLM.from_pretrained(MODEL_NAME)
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
```

```python
# 스페셜 토큰 값 확인
print('start:',tokenizer.cls_token_id)
print('end:',tokenizer.sep_token_id)
print('mask',tokenizer.mask_token_id)
```

```python
# BERT는 [MASK]라는 빈칸을 예측하는데 특화되어 있습니다.
text = 'Soccer is a really fun [MASK].'
inputs = tokenizer(text, return_tensors='tf')

inputs
```

```python
print(inputs['input_ids'])
```

```python
from transformers import FillMaskPipeline
pip = FillMaskPipeline(model=model, tokenizer=tokenizer)
```

```python
pip(text)[0]['sequence']
```

```python
pip('The Avengers is a [MASK].')
```

```python
for i in range(5):
    print(pip('i went to [MASK] this morning')[i]['sequence'])
```
