### **2) plotly import하기**



```python
import plotly.express as px
```

### **3) 데이터 불러오기**



```python
iris = px.data.iris()
iris.head()
```

```python
## gapminder 데이터는 국가별 년도별 life expectancy, GDP per capita, 그리고 population에 대한 데이터입니다.
gapminder = px.data.gapminder()
gapminder.head()
```

<br>

## **3. Scatter Plot**

```python
px.scatter(iris, x='sepal_width', y='sepal_length', color='species', size='petal_length')
```

## **4. Line Chart**


```python
df = gapminder.query("country=='Canada'")
df
```

```python
gapminder[gapminder['country']=='Canada']
```

* gapminder 데이터에서 country값이 'Canada'인 데이터만 df에 저장했습니다.

```python
px.line(df, x='year', y='lifeExp')
```

<br>

<br>

## **5. Box Plot**

box()는 상자 그림을 그리는 함수입니다.

```python
px.box(df, y='gdpPercap')
```

<br>

## **6. Histogram**

ㅤ**✔ㅤhistogram()**는 히스토그램을 그리는 함수입니다.
<br><br>
ㅤ**✔ㅤnbins** 파라미터로 히스토그램 빈의 개수를 설정할 수 있습니다.
<br><br>

```python
px.histogram(df, x='gdpPercap')
```

<br>
