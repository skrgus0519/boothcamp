# 활성화 함수 구현

### 활성화 함수 : 신호의 총 합을 출력 신호로 변환하는 함수

```python
import numpy as np
import matplotlib.pylab as plt
```

# Step Function

```python
# x내부 요소의 값이 음수이면, 0
# x내부 요소의 값이 양수이면, 1로 변경 후 반환
def step_function(x):
    y = np.array(x > 0).astype(int)

    return y

# 입력값(x)는 넘파이 배열.
X = np.arange(-5.0, 5.0, 0.1)
Y = step_function(X)
#print(Y)

plt.plot(X, Y)
# y축 범위 지정
plt.ylim(-0.1, 1.1)
plt.show()
```

# Sigmoid 함수

```python
def sigmoid(x):
    # np.exp(): x를 지수함수로 변경해주는 코드
    y = 1 / (1 + np.exp(-x))
    return y

Y = sigmoid(X)
# print(Y)

plt.plot(X,Y)
# y축 범위 지정
plt.ylim(-0.1, 1.1)
plt.show()
```

# ReLU 함수

```python
# np.maximum(a ,b) a와 b 각각의 요소를 비교하여
# 두 개의 배열 중 더 큰 값을 선택해 새로운 배열을 반환한다.

def relu(x):
    y = np.maximum(0, x)
    return y

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x,y)
# y축 범위 지정
plt.ylim(-0.5, 5.5)
plt.show()
```

# Hyperbolic Tangent

```python
def tahn(x):
    p_exp = np.exp(x)
    m_exp = np.exp(-x)

    return (p_exp - m_exp) / (p_exp + m_exp)

x = np.arange(-5.0, 5.0, 0.1)
y = tahn(x)

plt.plot(x, y)
plt.ylim(-1.1, 1.1)
plt.show()
```

# 활성화 함수 비교

```python
x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(x)
y2 = step_function(x)
y3 = relu(x)
y4 = tahn(x)

plt.plot(x, y1)
plt.plot(x, y2)
# plt.plot(x, y3)
plt.plot(x, y4)

plt.ylim(-1.1, 1.1)
plt.show()
```

## Softmax 함수
입력이 들어왔을 때, 모든 입력값에 대한 출력 확률을 반환하는 함수

```python
def softmax(x):
    exp = np.exp(x)
    sum_exp = np.sum(exp)
    return exp / sum_exp

x = np.array([2, 3, 1, 4])
y = softmax(x)

#for i in range(4):
#    print(x[i], "->" , y[i])
for i in range(len(x)):
    print(x[i], "->", y[i])

plt.pie(y, labels=y)
plt.show()
```
