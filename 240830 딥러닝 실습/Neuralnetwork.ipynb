# 순전파 구현

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
# 활성화 함수 정의
def sigmoid(x):
    return 1 / (1+np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def same(x):
    return x
```

# Neural Network 함수 정의

```python
import random as r
def init_network():
    """
    network = {}
    network['W1'] = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
    network['b1'] = np.array([0.1, 0.2, 0.3])
    network['W2'] = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.1]])
    network['b2'] = np.array([0.1, 0.2])
    network['W3'] = np.array([0.1, 0.3])
    network['b3'] = np.array([0.1])
    """
    network = {}
    network['W1'] = np.random.rand(2, 3)
    network['b1'] = np.zeros(3)
    network['W2'] = np.random.rand(3, 2)
    network['b2'] = np.zeros(2)
    network['W3'] = np.random.rand(2, 1)
    network['b3'] = np.zeros(1)
    return network

# 정의한 network, input
def forward(network, x):
    # 가중치, 편향 불러오기
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']

    # 행렬 곱셈을 통한 첫번째 Layer 계산
    a1 = np.dot(x, W1) + b1

    # 첫번째 노드의 활성화 함수로 sigmoid 함수 사용(현재 상태를 0~1 사이의 값으로 표시)
    z1 = sigmoid(a1)

    # 행렬 곱셈을 통한 두번째 Layer 계산
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    
    # 행렬 곱셈을 통한 세번째 Layer 계산
    a3 = np.dot(z2, W3) + b3
    z3 = relu(a3)

    # Output Layer 항등함수 사용(회귀)
    y = same(z3)
    return y

# 정의한 신경망 구조 가져오기
network = init_network()

# input layer
input = np.array([60, 90])
output = forward(network, input)
print(output) # 39.7점

target = 0.85
```

# 오차 계산

```python
def mean_squared_error(y, t):
    return np.sum((y-t)**2) / len(y)

def loss(x, t=target):
    y = forward(network, x)
    loss = mean_squared_error(y, t)

    return loss
```

```python
network = init_network()
epochs = 3000
lr = 0.01

for i in range(epochs):
    output = forward(network, input)

    if (output > target):
        # loss 대신 gradient 사용해야 함.
        network['W1'] -= loss(input) * lr
        network['b1'] -= loss(input) * lr
        network['W2'] -= loss(input) * lr
        network['b2'] -= loss(input) * lr
        network['W3'] -= loss(input) * lr
        network['b3'] -= loss(input) * lr
    else:
        network['W1'] += loss(input) * lr
        network['b1'] += loss(input) * lr
        network['W2'] += loss(input) * lr
        network['b2'] += loss(input) * lr
        network['W3'] += loss(input) * lr
        network['b3'] += loss(input) * lr

    print(f'{i+1}epochs loss:', loss(input))
    print(f'{i+1}epochs output:', output)
    print()
```
